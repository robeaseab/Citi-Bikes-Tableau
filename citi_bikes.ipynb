```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# create base df using first bike data csv
bike_data = "Resources/JC-201901-citibike-tripdata.csv"
bike_df = pd.read_csv(bike_data)
bike_df.tail()
```

```python
# collect additional 2019 csvs and create loop to add to bike df
csv_files = ["Resources/JC-201902-citibike-tripdata.csv",
            "Resources/JC-201903-citibike-tripdata.csv",
            "Resources/JC-201904-citibike-tripdata.csv",
            "Resources/JC-201905-citibike-tripdata.csv"]
for file in csv_files:
    add_data = file
    add_df = pd.read_csv(file)
    bike_df = bike_df.append(add_df)


bike_df.tail()   
    
```

```python
# separate start/stop date/time columns
start =  bike_df["starttime"].str.split(" ", n = 1, expand = True) 
start = start.rename(columns={0: 'start date', 1: 'start time'})

stop =  bike_df["stoptime"].str.split(" ", n = 1, expand = True) 
stop = stop.rename(columns={0: 'stop date', 1: 'stop time'})
stop.head()
```

```python
# append separate date and time columns
bike_df["start time"]= start["start time"]
bike_df["start date"]= start["start date"]

bike_df["stop time"]= stop["stop time"]
bike_df["stop date"]= stop["stop date"]

# drop original date time columns
bike_df = bike_df.drop("stoptime", axis = 1)
bike_df = bike_df.drop("starttime", axis = 1)

#explicitly state gender
bike_df['gender'] = bike_df['gender'].replace(to_replace= [0,1,2], value= ["unknown", "male", "female"])
bike_df.head()
```

```python
bike_df["age"] = 2019-bike_df["birth year"]
bike_df.head()
```

```python
bike_df.to_csv("Resources/output/citibike-tripdata-2019-all.csv")
```
